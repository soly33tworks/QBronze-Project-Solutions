We play a correlation game with classical bits. 

_We represent the basis state by using brackets such as $ \pstate{0} $ or $ \pstate{10} $._

### Create a python class called `CorrelationGame()`
Each instance of this class refers to a single probabilistic system. When an instance is created, it has a single bit that is in state $ \onehalf \pstate{0} + \onehalf \pstate{1} $.

This class allows us to add new bits and also remove the existing bit(s). 

The bits are enumerated starting from 0, and they are denoted $ b_0, b_1,\ldots,b_n $. The state of system is kept as a linear combination of the basis states, e.g., $ \onehalf \pstate{01100} + \onehalf \pstate{11110} $.


### The methods

1. `add_a_new_bit(state_value=-1)`: Add a new classical bit in the specified state (pick the state randomly if state_value is -1) to the system. For example, if the system is in state $ \onehalf \pstate{0} + \onehalf \pstate{1} $ and the method `add_a_new_bit(1)` is called, then the state of the system will be $ \onehalf \pstate{01} + \onehalf \pstate{11}  $. Remark that the index of the new bit will be the number of previous bits.

1. `print_state()`: Print the state of the system as a linear combination of the basis states. 

1. `print_state_vector()`: Print the state of the system as a list corresponding to the state vector.

1. `not(index_of_bit)`: Apply the NOT operator on the specified bit.

1. `cnot(control_bit,target_bit)`: Apply a CNOT operator on the specified bits.

1. `random_cnot()`: Apply a CNOT operator on two randomly picked bits.

1. `is_correlated(index_of_bit)`: Determine whether the specified bit is part of the correlation.

1. `uncorrelated_bits()`: Return the list of uncorrelated bits.

1. `correlated_bits()`: Return the list of correlated bits.

1. `create_correlations()`: For each uncorrelated bit, say $ b_j $, a bit among the correlated bits is randomly picked, say $b_i$, and CNOT(i,j) is applied.

1. `remove_an_uncorrelated_bit()`: Do nothing if there is a single bit in the system. Randomly pick an uncorrelated bit, remove it from the system, update the state vector accordingly, and return the value of the removed bit.

1. `remove_uncorrelated_bits()`: Do nothing if there is a single bit in the system. Remove all uncorrelated bits from the system and update the state vector accordingly as long as there is more than one bit.

```python
"1-Linear algebra functions"

# Multiply matrix A with vector v
def mat_vec_mul(A, v):
    Av = []
    for row in A:
        Av.append(sum([row[i]*v[i] for i in range(len(v))]))
    return Av

# Multiply matrix M with matrix N
def matmul(M, N):
    C = []
    for i in range(len(M)):
        C.append([])
        for j in range(len(N)):
            val = sum([M[i][m]*N[m][j] for m in range(len(M[0]))])
            C[i].append(val)
    return C

# Tensor multiplication of matrix A with matrix B
def tensor_mul_mat(A, B):
    cols = len(A[0])*len(B[0])
    rows = len(A)*len(B)

    AB=[]
    for i in range(rows):
        AB.append([])
        for j in range(cols):
            AB[i].append(0)

    for i in range(len(A)):
        for j in range(len(A[i])):
            for k in range(len(B)):
                for l in range(len(B[k])):
                    AB[len(B)*i+k][len(B[0])*j+l] = A[i][j]*B[k][l]
    return AB


"2-Helper functions for creating classical states"

# Expands the current basis by 1 bit (e.g. [0 1] --> [00 01 10 11])
def create_basis(curr_basis):
    bits = ["0", "1"]
    basis_states = []
    for basis in curr_basis:
        for i in range(2):
            basis_states.append(basis+bits[i])
    return basis_states

# Expands & reorders the state vector itself to represent all basis (to show 0 valued states and print in correct order)
def organize_state(n, current_state):
    basis_states = [""]
    for bit in range(n): # Create a basis state with n bits
        basis_states = create_basis(basis_states)
    
    state = {}
    for basis in basis_states:
        if basis in current_state.keys():
            state[basis] = current_state[basis]
        else:
            state[basis] = 0
    
    return state

# Deletes states with 0 probability
def get_nonzero_states(state):
    nonzero_state = {}
    for key in state.keys():
        if state[key] != 0:
            nonzero_state[key] = state[key]
    return nonzero_state
```

```python
from random import randrange, sample

X = [[0, 1],
     [1, 0]]

I = [[1, 0],
     [0, 1]]

class CorrelationGame:
    
    def __init__(self):
        self.n = 1
        self.state = {"0":1/2, "1":1/2}
        self.state_vector = list( self.state.values() )
    
    def add_a_new_bit(self, state_value=-1):
        if state_value in [-1, 0, 1]:
            if state_value == -1:
                state_value = randrange(2)
            
            new_state = {}
            for key in self.state.keys():
                new_state[key+ str(state_value)] = self.state[key]
            self.state = new_state
            self.n += 1
            
            self.state = organize_state(self.n, self.state)
            self.state_vector = list( self.state.values() )
            
        else:
            print("Invalid value, select 0 or 1 (-1 for random)")
    
    def print_state(self):
        #print(self.state)
        print(get_nonzero_states(self.state))
    
    def print_state_vector(self):
        print(self.state_vector)
    
    def x(self, index_of_bit):
        if self.n > index_of_bit:
            if self.n == 1:
                self.state_vector = mat_vec_mul(X, self.state_vector)

            else:
                big_X = [[1]]
                for bit in range(self.n):
                    if index_of_bit == bit:
                        big_X = tensor_mul_mat(big_X, X)
                    else:
                        big_X = tensor_mul_mat(big_X, I)
                
                self.state_vector = mat_vec_mul(big_X, self.state_vector)
                    
            for i, key in enumerate(self.state.keys()):
                self.state[key] = self.state_vector[i]
                
            self.state_vector = list( self.state.values() )
        else:
            print("Index doesn't exist")
    
    def cx(self, control_bit, target_bit):
        if (self.n >= 2) and (self.n > control_bit) and (self.n > target_bit) and (control_bit != target_bit):
            new_state = {}
            
            for i, key in enumerate(self.state.keys()):
                if key[control_bit] == "1":
                    new_key = key                    
                    new_bit = str(1- int(key[target_bit]))
                    new_key = new_key[:target_bit] + new_bit + new_key[target_bit+1:]
                    new_state[new_key] = self.state[key]
                else:
                    new_state[key] = self.state[key]
                    
            self.state = organize_state(self.n, new_state)
            self.state_vector = list( self.state.values() )
        else:
            print("Invalid indexes")
    
    def random_cnot(self):
        control_bit, target_bit = sample(range(self.n), 2)
        self.cx(control_bit, target_bit)
    
    def is_correlated(self, index_of_bit):
        nonzero_state = get_nonzero_states(self.state)
        
        bits = [key[index_of_bit] for key in nonzero_state.keys()]
        is_correlated = True
        if (bits.count("0") > 1) or (bits.count("1") > 1):
            is_correlated = False
        return is_correlated
    
    def uncorrelated_bits(self):
        uncorr_bits = []
        for bit in range(self.n):
            if not self.is_correlated(bit):
                uncorr_bits.append(bit)
        return uncorr_bits
    
    def correlated_bits(self):
        uncorr_bits = self.uncorrelated_bits()
        corr_bits = [value for value in list(range(self.n)) if value not in uncorr_bits]
        return corr_bits
    
    def create_correlations(self):
        if self.n != 1:
            uncorr_bits = self.uncorrelated_bits()
            corr_bits = self.correlated_bits()
            
            for bit in uncorr_bits:
                random_bit = corr_bits[randrange(len(corr_bits))]
                self.cx(bit, random_bit)
                corr_bits.append(bit)
    
    def remove_an_uncorrelated_bit(self):
        if self.n != 1:
            uncorr_bits = self.uncorrelated_bits()
            if len(uncorr_bits) > 0:
                random_bit = uncorr_bits[randrange(len(uncorr_bits))]

                self.n -= 1
                new_state = organize_state(self.n, current_state={})

                for key in self.state.keys():
                    new_key = key[:random_bit] + key[random_bit+1:]
                    new_state[new_key] += self.state[key]
                self.state = new_state

                self.state = organize_state(self.n, self.state)
                self.state_vector = list( self.state.values() )

                return random_bit
            else:
                print("All bits are correlated")
                return None
        else:
            print("Only 1 bit remains")
            return None
    
    def remove_uncorrelated_bits(self):
        uncorr_bits = self.uncorrelated_bits()
        if len(uncorr_bits) > 0:
            for i in range(len(uncorr_bits)):
                removed_bit = self.remove_an_uncorrelated_bit()
                if removed_bit is None:
                    break
        else:
            print("All bits are correlated")
```
