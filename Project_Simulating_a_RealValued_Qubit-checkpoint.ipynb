We play with a single qubit living on the unit circle.

- _All angles are in radian._
- _Please do not use any quantum programming library or any scientific python library such as `NumPy`._
- _You can use python module `math` and python interface `matplotlib.pyplot`._

### Create a python class called `SingleQubit(theta=0)`

When an instance is created, the quantum state of qubit is set to the specified angle _theta_. If no angle is specified by the user, then the initial state of the qubit will be $\ket{0}$. 

### The basic methods

1. `read_state()`: Return the current quantum state of the qubit.

1. `rotation(theta)`: Update the current quantum state by applying the rotation operator with the angle _theta_.

1. `reflection(theta)`: Update the current quantum state by applying the reflection operator having a reflection axis with angle _theta_.

1. `draw_state()`: Draw the current quantum state as an arrow on the unit circle.

1. `draw_all_states()`: Draw all the quantum states visited on the unit circle starting from the initial one by indicating their visited orders.

1. `reflect_and_draw(theta)`: Draw the unit circle, the current quantum state, the reflection line of the reflection operator, apply the method `reflection(theta)` to the current quantum state, and draw the new quantum state.

1. `prob()`: Return the probabilities of observing state 0 and state 1 if the qubit is measured at this moment.

1. `measure(number_of_shots)`: Simulate the measurement operator _number_of_shots_ times based on the current quantum state, and then return the number(s) of the observed 0’s and 1’s (remark that only a single state is observed after each measurement, i.e., either 0 or 1).

### The extra methods

9. `change_basis(theta)`: Return an error if the basis will remain the same. Change the basis of qubit with the unit vectors $ \ket{v_1} $ and $ \ket{v_2} $ such that 
    - the angles of $ \ket{v_1} $ and $ \ket{v_2} $ are $ \theta $ and $ \theta + \frac{\pi}{2} $ in the standard basis, respectively, and 
    - $\ket{v_1}$ becomes the new $ \ket{0} $ and $ \ket{v_2} $ becomes the new $ \ket{1} $; and then,   
update all visited quantum states until now respectively, i.e, each quantum state in the standard basis should be shifted by $ -\theta $.

2. `draw_state_in_both_basis()`: Return an error if the basis has not changed. Draw two separete unit circles where the basis are the current basis and the previous basis, and then draw the current quantum state as an arrow on each unit circle.

3. `prob_in_both_basis()`: Return an error if the basis has not changed. Return the probabilities of observing state 0 and state 1 in the current and previous bases.

4. `take_back_basis_change()`: Do nothing if there is no basis change. Return back to the previous basis by deleting the latest call of `change_basis()` from the basis history. On the other hand, any quantum state visited after the latest call of `change_basis()` must be kept.

```python
"1-Linear algebra functions"

# Multiply matrix A with vector v
def mat_vec_mul(A, v):
    Av = []
    for row in A:
        Av.append(sum([row[i]*v[i] for i in range(len(v))]))
    return Av
    
"2-Plot related functions (taken from official QBronze repo)"
def draw_axes():
    points = [ [1.2,0], [0,1.2], [-1.2,0], [0,-1.2] ] # dummy points for zooming out
    arrows = [ [1.1,0], [0,1.1], [-1.1,0], [0,-1.1] ] # coordinates for the axes
    for p in points: 
        plt.plot(p[0],p[1]+0.1) # drawing dummy points
    for a in arrows: 
        plt.arrow(0,0,a[0],a[1],head_width=0.04, head_length=0.08) # drawing the axes
    
    plt.text(1.05,0.05,"|0>")
    plt.text(0.05,1.05,"|1>")
    plt.text(-1.2,-0.1,"-|0>")
    plt.text(-0.2,-1.1,"-|1>")

def draw_unit_circle(init_fig = True):
    if init_fig:
        plt.figure(figsize=(7,7), dpi=60)
    draw_axes()
    unit_circle = plt.Circle((0,0),1,color='black',fill=False)
    plt.gca().add_patch(unit_circle)
```

```python
from math import pi, cos, sin, sqrt
from matplotlib import pyplot as plt
from random import choices

class SingleQubit():
    
    def __init__(self, theta=0):
        self.state = [cos(theta), sin(theta)] # State is initialized at θ degrees
        self.past_states = [self.state]
        self.basis_angle = [0] # We start with basis θ=0°
    
    # BASIC METHODS
    def read_state(self):
        return self.state
    
    def rotation(self, theta):
        R = [[cos(theta), -sin(theta)], # Rotation matrix
             [sin(theta), cos(theta)]]
        
        self.state = mat_vec_mul(R, self.state)
        self.past_states.append(self.state)
    
    def reflection(self, theta):
        Ref = [[cos(2*theta), sin(2*theta)], # Reflection matrix
               [sin(2*theta), -cos(2*theta)]]
        
        self.state = mat_vec_mul(Ref, self.state)
        self.past_states.append(self.state)
    
    def draw_state(self, init_fig = True):
        draw_unit_circle(init_fig)
        
        x, y = self.state
        plt.arrow(0,0, x*0.92,y*0.92, head_width=0.04,head_length=0.08, color="blue")
        plt.text(x*1.15, y*1.15, '|v'+str(len(self.past_states))+'⟩')
    
    def draw_all_states(self):
        draw_unit_circle()
        
        for i, state in enumerate(self.past_states):
            x, y = state
            plt.arrow(0,0, x*0.92,y*0.92, head_width=0.04,head_length=0.08, color="blue")
            plt.text(x*1.15, y*1.15, '|v'+str(i+1)+'⟩')
    
    def reflect_and_draw(self, theta):
        draw_unit_circle()
        
        # Draw, reflect and draw again
        self.draw_state(init_fig = False)
        self.reflection(theta)
        self.draw_state(init_fig = False)
        
        slope = sin(theta)/cos(theta)
        x = 1.1/sqrt(slope**2+1)
        y = x*slope
        dx = -2*x
        dy = -2*y
        plt.arrow(x,y, dx,dy,linestyle='dotted',color='red')
    
    def prob(self):
        p0, p1 = self.state
        return {"0": p0**2, "1": p1**2}
    
    def measure(self, number_of_shots):
        probabilities = self.prob()
        weights = [probabilities["0"], probabilities["1"]]
        
        results = choices(["0", "1"], weights=weights, k=number_of_shots)
        counts = {"0": results.count("0"), "1": results.count("1")}
        return counts
    
    # EXTRA METHODS
    def change_basis(self, theta):
        curr_angle = self.basis_angle[-1] # Basis angle is not cumulative (its wrt 0°)
        if theta != curr_angle:
            
            delta_theta = theta - curr_angle
            self.basis_angle.append(theta)
            R = [[cos(-delta_theta), -sin(-delta_theta)],
                 [sin(-delta_theta), cos(-delta_theta)]]
        
            self.state = mat_vec_mul(R, self.state)
            past_states = [mat_vec_mul(R, state) for state in self.past_states]
            self.past_states = past_states

        else:
            print("The basis is the same")
    
    def draw_state_in_both_basis(self):
        if len(self.basis_angle) >= 2: # Execute if we have a previous basis angle
            
            draw_unit_circle()
            delta_theta = self.basis_angle[-1] - self.basis_angle[-2] # Previous change in angle
            R = [[cos(delta_theta), -sin(delta_theta)], # Matrix to reverse previous change
                 [sin(delta_theta), cos(delta_theta)]]
            x_prev, y_prev = mat_vec_mul(R, self.state)
            x, y = self.state

            plt.arrow(0,0, x_prev*0.92,y_prev*0.92, head_width=0.04,head_length=0.08, color="blue")
            plt.text(x_prev*1.15, y_prev*1.15, 'previous basis')
            plt.arrow(0,0, x*0.92,y*0.92, head_width=0.04,head_length=0.08, color="blue")
            plt.text(x*1.15, y*1.15, 'current basis')
        else:
            print("Previous basis doesn't exist")
    
    def prob_in_both_basis(self):
        if len(self.basis_angle) >= 2:

            delta_theta = self.basis_angle[-1] - self.basis_angle[-2] # Previous change in angle
            R = [[cos(delta_theta), -sin(delta_theta)],
                 [sin(delta_theta), cos(delta_theta)]]
            x_prev, y_prev = mat_vec_mul(R, self.state)
            
            prob_prev = {"0": x_prev**2, "1": y_prev**2}
            prob_current = self.prob()
            
            return prob_current, prob_prev

        else:
            print("Previous basis doesn't exist")
            return None, None
    
    def take_back_basis_change(self):
        if len(self.basis_angle) >= 2:
            self.change_basis(self.basis_angle[-2])
            # Delete the last two angles. One is the previous angle and one was added just now with change_basis automatically.
            self.basis_angle = self.basis_angle[:-2]
            
        else:
            print("No basis change was made")
```

### Testing

```python
# Intialization
q = SingleQubit(pi/6)
print("Probabilities:",q.prob())
q.draw_state()
```

```python
# Reflection and rotation
q.reflect_and_draw(pi/5)
q.rotation(pi/2)
q.draw_state()
```

```python
# Change of basis
q.change_basis(pi/2)
q.draw_state_in_both_basis()
p_curr, p_prev = q.prob_in_both_basis()
print("Current basis probabilities (θ=0°):",p_curr)
print("Previous basis probabilities (θ=90°):",p_prev)
```

```python
# Take back basis change etc.
q.take_back_basis_change()
q.draw_state_in_both_basis() # Will give error since we took back change
q.draw_state()
q.draw_all_states()
```
