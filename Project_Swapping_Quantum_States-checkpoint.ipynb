### Create a python class called `SwappingGame()`

<img src="images/seven_qubits.png" width="200px" align="right" style="margin:20px;">

We have 7 qubits placed on a graph as shown on the right: $ q_{-3} $, $q_{-2}$, $q_{-1}$, $q_{0}$, $q_{+1}$, $q_{+2}$, and $q_{+3} $.

We are allowed to implement only $ CNOT $ operators, and a $ CNOT $ operator is directly applicable to the neighbor qubits, i.e., two qubits are neighbors if  there is an edge (line) between them. For example a $ CNOT $ operator can be applied to the qubits $q_{-3}$ and $ q_{-1} $ or $q_0$ and $q_{+1}$ or $q_{+1}$ and $q_{+2}$, but a $ CNOT $ operator cannot be directly applied to the qubits $q_{-1}$ and $q_{+1}$. 

#### The quantum system

These 7 qubits form a single quantum system. Then, the quantum state of the system is represented by a $2^7$-dimensional vector (list) and so each quantum operator is represented by a ($2^7 \times 2^7$)-dimensional matrix (double dimensinal list).

The system should be simulated as a whole:
- The simulation of the whole system should not be done by separately simulating each qubit or pairs of qubits and then combine them.
- The aim here is to experience the simultion of generic quantum systems.
    
There are 6 pairs of the neighbor qubits, and so 12 different $ CNOT $ operators should be applied to this system:
- $ CNOT(i,j) $ is the $ CNOT $ operator applied to the control qubit $ q_i $ and the target qubit $q_j $.

#### Initialization

When an instance of the class is created, an angle, say $\theta$, is randomly picked in the interval $(0,\pi/6)$. 

The quantum state of the qubit $ q_i $, say $\ket{v_i}$, is set to the state $ \myvector{\cos \theta \\ \sin \theta} $.

### The main method `swap(*,*)`

This class has one main method called `swap(i,j)` for $i,j \in \{-3,\ldots,3\} $, which swaps the states of the qubits $ q_i $ and $ q_j $.

#### The auxiliary method `swap_neighbors(*,*)`

An auxiliary method called `swap_neighbors(i,j)` is defined to swap the states of the neighbor qubits $ q_i $ and $ q_j $. This method gives a warning message whenever the qubits $ q_i $ and $ q_j $ are not neighbors.

Whenever two qubits are not neighbor, we will use a series of `swap_neighbors(*,*)` to the qubits on a path connecting them. For example, the states of the qubits $ q_{+2} $ and $ q_{-3} $ can be swaped as described below. Assume that the state of $ q_{i} $ is $ \ket{v_i} $ before starting.
-  After `swap_neighbors(2,1)`, the states of $ q_{+2} $ and $ q_{+1} $ are $ \ket{v_1} $ and $ \ket{v_2} $, respectively.
-  After `swap_neighbors(1,0)`, the states of $ q_{+1} $ and $ q_{0} $ are $ \ket{v_0} $ and $ \ket{v_2} $, respectively.
-  After `swap_neighbors(0,-1)`, the states of $ q_{0} $ and $ q_{-1} $ are $ \ket{v_{-1}} $ and $ \ket{v_2} $, respectively.
-  After `swap_neighbors(-1,-3)`, the states of $ q_{-1} $ and $ q_{-3} $ are $ \ket{v_{-3}} $ and $ \ket{v_2} $, respectively.
\
Now, the current state of qubit $q_{-3}$ is $ \ket{v_{2}} $ as desired. 
-  After `swap_neighbors(-1,0)`, the states of $ q_{-1} $ and $ q_{0} $ are $ \ket{v_{-1}} $ and $ \ket{v_{-3}} $, respectively.
-  After `swap_neighbors(0,1)`, the states of $ q_{0} $ and $ q_{1} $ are $ \ket{v_{0}} $ and $ \ket{v_{-3}} $, respectively.
-  After `swap_neighbors(1,2)`, the states of $ q_{1} $ and $ q_{2} $ are $ \ket{v_{1}} $ and $ \ket{v_{-3}} $, respectively.
\
And now, the current state of qubit $ q_2 $ is $ \ket{v_{-3}} $.
\
Besides, the state of each other qubit is the same as in the beginning.


#### Implementing `swap_neighbors(*,*)`

The swap operator between two neighbor qubits, say $q_i$ and $q_j$ is implemented by three $ CNOT $ operators:

    CNOT(i,j)
    CNOT(j,i)
    CNOT(i,j)

*You may verify the correctness of this method by using the basis states $ \ket{00} $, $ \ket{01} $, $ \ket{10} $, and $ \ket{11} $.*

This method multiplies the current quantum state with the three matrices representing these $ CNOT $ operators.

```python
"1-Linear algebra functions"

# Tensor multiplication of vector a with vector b
def tensor_mul_vec(a, b):
    ab = []
    for ai in a:
        for bi in b:
            ab.append(ai*bi)
    return ab


"2-Helper functions for creating states"

# Expands the current basis by 1 bit (e.g. [0 1] --> [00 01 10 11])
def create_basis(curr_basis):
    bits = ["0", "1"]
    basis_states = []
    for basis in curr_basis:
        for i in range(2):
            basis_states.append(basis+bits[i])
    return basis_states

# Expands & reorders the state vector itself to represent all basis (to show 0 valued states and print in correct order)
def organize_state(n, current_state):
    basis_states = [""]
    for bit in range(n): # Create a basis state with n bits
        basis_states = create_basis(basis_states)
    
    state = {}
    for basis in basis_states:
        if basis in current_state.keys():
            state[basis] = current_state[basis]
        else:
            state[basis] = 0
    
    return state

# Deletes states with 0 probability
def get_nonzero_states(state):
    nonzero_state = {}
    for key in state.keys():
        if state[key] != 0:
            nonzero_state[key] = state[key]
    return nonzero_state

# Creates the state vector by applying tensor multiplication to the qubits
def get_state_vector(qubits):
    n = len(qubits)
    state_vector = qubits[0]
    for i in range(n-1):
        state_vector = tensor_mul_vec(state_vector, qubits[i+1])
        
    return state_vector

# Creates the state using the state vector
def get_state_from_statevec(state_vector, n=7):
    basis_states = [""]
    for bit in range(n): # Create a basis state with n bits
        basis_states = create_basis(basis_states)
    
    state = {}
    for i, basis in enumerate(basis_states):
        state[basis] = state_vector[i]
        
    return state


"3-Path related function"

# Finds the path of neighboring nodes between node i and j
def find_path(i, j): # i!=j
    if i>j:
        increasing = False
    elif i<j:
        increasing = True

    diagram = [[2], [2], [0,1,3], [2,4], [3,5,6], [4], [4]] # The map of the diagram showing connections
    path = [i]
    pos = i
    choices = diagram[pos]
    while pos != j:
        if len(choices) == 1:
            pos = choices[0]
            choices = diagram[pos]
            path.append(pos)
        elif j in choices:
            pos = j
            choices = diagram[pos]
            path.append(pos)
        else:
            if increasing:
                choices = [e for e in choices if e>pos]
            else:
                choices = [e for e in choices if e<pos]
    
    path_go = path
    path_back = list(reversed(path[:-1]))
    return path_go, path_back
```

```python
from random import randrange
from math import pi, cos, sin

class SwappingGame():
    
    def __init__(self):
        self.n = 7
        self.qubits = []
        for i in range(self.n):
            theta = (randrange(3000)/3000)*pi/6
            q = [cos(theta), sin(theta)]
            self.qubits.append(q)
        
        self.state_vector = get_state_vector(self.qubits)
        self.state = get_state_from_statevec(self.state_vector, n=self.n)
    
    def print_state(self):
        #print(self.state)
        print(get_nonzero_states(self.state))
    
    def print_state_vector(self):
        print(self.state_vector)
    
    # Its difficult to write the tensor form of the CNOT gate applied on the whole circuit when ∣i-j∣ > 1.
    # Since we are swapping neighbors, it is guaranteed that ∣i-j∣ <= 2 for this map, which makes it simpler.
    # But I did it bitwise instead of using matrix vector multiplication since it was easier.
    def cx(self, control_bit, target_bit):
        if (self.n >= 2) and (self.n > control_bit) and (self.n > target_bit) and (control_bit != target_bit):
            new_state = {}
            
            for i, key in enumerate(self.state.keys()):
                if key[control_bit] == "1":
                    new_key = key                    
                    new_bit = str(1- int(key[target_bit]))
                    new_key = new_key[:target_bit] + new_bit + new_key[target_bit+1:]
                    new_state[new_key] = self.state[key]
                else:
                    new_state[key] = self.state[key]
                    
            self.state = organize_state(self.n, new_state)
            self.state_vector = list( self.state.values() )
        else:
            print("Invalid indexes")
    
    def swap(self, i, j):
        i, j = i+3, j+3
        if i!=j:
            path_go, path_back = find_path(i, j) # Get the correct order of swapping the neighbors
            
            for step in range(len(path_go)-1):
                self.swap_neighbors(path_go[step], path_go[step+1])
            
            for step in range(len(path_back)-1):
                self.swap_neighbors(path_back[step], path_back[step+1])            

        else:
            print("Same node")
    
    def swap_neighbors(self, i, j):
        diagram = [[2], [2], [0,1,3], [2,4], [3,5,6], [4], [4]] # The map of the diagram showing connections
        
        if j in diagram[i]:
            # Swap using CNOT gates
            self.cx(i, j)
            self.cx(j, i)
            self.cx(i, j)
            
            # Manually swap on the qubits list for later verification
            val1, val2 = self.qubits[i], self.qubits[j]
            self.qubits[i] = val2
            self.qubits[j] = val1
        else:
            print("Nodes are not neighbors")
```

### Verify the calculations

One way to verify the correctness of the operators is to keep the current quantum state of each qubit separately in a list, say `list_of_states`. Then, after each swap operator on the neighbors, the state of system is compared with the tensor product of the states of qubits in the appropriate order.
